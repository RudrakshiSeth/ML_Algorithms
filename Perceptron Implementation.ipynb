```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def plot_data(mistake_list):
    plt.plot(mistake_list)
    plt.show()


def implement_preceptron(mistake_list,passes,X,y,w,b):
    for t in range(passes):
        mistake = 0
        for i in range(len(X)):
            new_prediction = y[i]*(np.dot(X[i],w)+b)
            if(new_prediction<=0):
                w= w+(y[i]*X[i])
                b = b+y[i]
                mistake = mistake+1
        mistake_list.append(mistake)
    return mistake_list

def main():
#inputing data sets
    X = pd.read_csv('spambase_X.csv',header=None).values.T
    y = pd.read_csv('spambase_y.csv',header=None).values
#intializing required variables#
    passes=500
    w=np.zeros(len(X[0]))
    mistake_list=[]
    mistake = 0
    b=0
    

    mistake_list=implement_preceptron(mistake_list,passes,X,y,w,b)
    plot_data(mistake_list)

if __name__=="__main__": 
    main()


```
